### Datos de tiempo

```python
df_meteorites['year']
```

```python
## Transformando en tipo tiempo

pd.to_datetime(
df_meteorites['year'],
errors = 'coerce',           #si encuentra un formato no fecha lodistingue y coloca algo diferente
format = '%m/%d/%Y %H:%M:%S %p'   #%p para ma√±ana tarde
)
```

Ahora, colocando el formato establecido anteriormente en el DataFrame:

```python
df_meteorites['year'] =  pd.to_datetime(
df_meteorites['year'],
errors = 'coerce',
format = '%m/%d/%Y %H:%M:%S %p'
)
```
