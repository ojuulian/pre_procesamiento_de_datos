```python
df_pob = pd.read_csv('1.csv',header=2)   #skiprows=4
df_pob
```

```python
list(df_pob)                
#para obtener columnas y proceder des pues a borrar           

df_pob = df_pob.drop(columns=['Country Code', 'Indicator Name','Indicator Code',  '1960',  '1961',    '1962', '1963', '1964', '1965', '1966', '1967', '1968', '1969', '1970', '1971', '1972', '1973', '1974', '1975', '1976',
 '1977', '1978', '1979', '1980', '1981', '1982', '1983', '1984', '1985', '1986', '1987', '1988', '1989', '1990', '1991', '1992', '1993', '1994',
 '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012',
 '2013', '2014', 'Unnamed: 64'])

df_pob
```

```python
df = pd.melt(df_pob, id_vars=['Country Name'], var_name='year', value_name='pob')
df = df.sort_values(by=['year','Country Name'])
df
```

```python
pd.options.display.float_format = '{:,.1f}'.format
```

```python
df_pob = df
df_pob 
```

```python
df_pob['year'] = pd.Categorical(df_pob['year'].apply(str))
#df['year']=df['year'].astype('category')
```

```python
df_pob.dtypes
```

```python
idx_filtro = df_pob['Country Name'].isin(['Aruba','Colombia'])   #genera variable boleana que permitira filtrar
idx_filtro
```

```python
df_sample = df_pob[idx_filtro]
df_sample
```

A continuación segeneran nuevos índices al DataFrame (En estecaso se establecen dos índices **Country Name** y **year**) este tipo de indices permite estructurar los datos para un mejor análisis.

```python
df_sample = df_sample.set_index(['Country Name','year']).sort_index()
df_sample
```

```python
df_sample.loc['Colombia',:]    #llamando los datos de Colombia
```

```python
df_sample.loc['Colombia',:].loc['2016',:]   #Llamando los datos del 2016
```

```python
df_sample.xs(['Aruba'])   #Con la función xs se pueden seleccionar varios índices
```

```python
df_sample.xs(['Aruba','2018'])  #seleccionando varios índices
```

```python
df_sample.xs('2018', level='year')
```
