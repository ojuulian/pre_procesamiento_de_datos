<h2 align="center">AGREGATION</h2>

A continuación se muestra un conjunto de datos pertenecientes a un restaurante, en este se muestra los detalles de la factura del cliente. Entre las características se destaca:

* **total_bill**: Precio de la factura (puede ser del plato).
* **tip**: Propina.
* **sex**: Sexo del cliente.
* **smoker**: Si fuma o no fuma.
* **day**: Dia de la semana.

```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.__version__
```

```python
pd.options.display.float_format = '{:,.3f}'.format
```

```python
df=sns.load_dataset('tips')
df
```

```python
df.describe(include='all')
```

Se puede observar que el costo promedio de un plato esta alrededor de 19 dólares, la propina alrededor de 3 USD. En cuanto a las variables categoricas se logra observar que:

* Existen dos variables de sexo.
* Existen dos opciones de respuesta en cuanto a fumar y no fumar.
* Existe registrado 4 dias de la semana..


```python
df['day'].value_counts()
```

```python
df['day'].value_counts()  /  df['day'].value_counts().sum()   * 100   #observando conteos como porcentajes.
```

```python
df.groupby('sex').mean()
```

La propina de los hombres es alrededor de 2.6USD y el de las mujeres de 2.4USD. Además los platos comprados por los hombres es de 20USD y el de las mujeres de 18USD.

### Porcentaje de propina

Considerando la especialidad del plato y su precio se determina un porcentaje de propina con respecto al valor dela factura:

```python
df['prct_tip']=df['tip']/df['total_bill']
df
```

```python
df.groupby('sex').mean()
```

A qui se puede destacar que los hombres y mujeres aportan un 15% del valor de sus pagos y lo destinan a propinas.

```python
df.groupby('sex').median()
```

```python
df.groupby('sex')[['prct_tip']].describe()
```

```python
df.groupby('sex')[['prct_tip','total_bill']].describe()
```

```python
#función paa convertir euros a dolares

def mean_eur2usd(x):
    return np.mean(x)*1.12
```

```python
mean_eur2usd([100,102])
```

```python
df.groupby('sex')[['total_bill','prct_tip']].apply(mean_eur2usd)   #aplicando una función con apply
```

```python
df.groupby(['sex','time'])[['total_bill','prct_tip']].apply(mean_eur2usd)     #clasificando por sexo y por tiempo
```

```python
df.groupby(['sex','time'])[['total_bill','prct_tip']].apply(np.std) 
```

### Crreando función lambda

```python
df.groupby(['sex','time'])[['total_bill','prct_tip']].apply(lambda x:  np.mean(x)*1.12) 
```

#### colocando mas deuna función usar aggregate (también agg)

```python
df.groupby(['sex','time'])[['total_bill','prct_tip']].aggregate([np.mean, np.max])
```

```python
df.groupby(['sex','time'])[['total_bill','prct_tip']].agg([np.mean, np.max])
```

```python
nuev_dict={'tip':[min, max], 'total_bill':[np.mean, mean_eur2usd]}    #este diccionario tiene un conjunto de funciones
nuev_dict
```

```python
df.groupby(['sex','time'])[['total_bill','tip']].agg(nuev_dict)
```

### CREANDO FILTRO POR CATEGORIAS CON MEDIA MAYOR A 40

```python
def f_filter(x):
    return mean_eur2usd(x['total_bill'].mean()) > 20
```

```python
df_filtered = df.groupby(['sex','time']).filter(f_filter)
df_filtered
```

anteriormente se elimino todas las categorias donde el costo promedio del plato en un determinado grupo sea mayor a 20

```python
df_filtered.groupby(['sex','time']).count()
```

anteriormente se muestra 0 en las mujeres en hora del almuerzo

generando intervalos en variables numéricas

```python
pd.cut(df['total_bill'],bins=3)
```

```python
pd.cut(df['total_bill'],bins=3).unique()
```

No obstante, se crean categorias con  respecto a la variable valor_factura donde existen 140 personas en el intervalo (3.022, 18.983], 88 personas en el intervalo de (18.983, 34.897] dolares y 16 e el intervalo (34.897, 50.81].

Definiendo el ancho de los intervalos con:

```python
pd.cut(df['total_bill'],bins=[3,18,35,60]).value_counts()
```

```python
df['bin_total']=pd.cut(df['total_bill'],bins=[3,18,35,60])
df
```

```python
df.groupby(['time','bin_total'])[['ones']].count()
```

```python
df.groupby(['time','bin_total'])[['ones']].count().groupby(level=0).apply(

lambda x:
    x/ x.sum()*100
)
```
